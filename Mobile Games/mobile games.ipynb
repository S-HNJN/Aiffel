```python
df['version'].value_counts()
```

```python
print(df.groupby('version')['retention_1'].mean()*100)
a_retention = df.groupby('version')['retention_1'].mean()['gate_30']
b_retention = df.groupby('version')['retention_1'].mean()['gate_40']
a_cnt = df[df['version']=='gate_30'].shape[0]
b_cnt = df[df['version']=='gate_40'].shape[0]
a_effected = round(a_retention * a_cnt)
b_effected = round(b_retention * b_cnt)
a_failed = a_cnt - a_effected
b_failed = b_cnt - b_effected
```

```python
a= [a_effected, a_failed]
b= [b_effected, b_failed]
conversion_table = np.array([a, b])
conversion_table = np.array([a, b])
```

```python
chi2_stat, p_value, dof, expected = stats.chi2_contingency(conversion_table)
```

```python
print(f"Chi-Square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.10f}")

alpha = 0.05  # 유의수준 5%
if p_value < alpha:
    print("A/B 프로모션 간 전환율 차이가 유의미합니다.")
else:
    print("A/B 프로모션 간 전환율 차이가 유의미하지 않습니다.")
```

```python
import scipy.stats as stats

a_sum = np.array(df[df['version'] == 'gate_30']['sum_gamerounds'])
b_sum = np.array(df[df['version'] == 'gate_40']['sum_gamerounds'])

# 대응 표본 t-검정 수행
t_stat, p_value = stats.ttest_ind(a_sum, b_sum, equal_var=False)

# 결과 출력
print(f"t-statistic: {t_stat:.4f}")
print(f"p-value: {p_value:.10f}")  # 소수점 10자리까지 표시

# 해석
alpha = 0.05  # 유의수준 5%
if p_value < alpha:
    print("유의미합니다.")
else:
    print("유의미하지 않습니다.")
```
